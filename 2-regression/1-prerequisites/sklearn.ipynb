## Import library

```python
import matplotlib.pyplot as plt 
import numpy as np 
from sklearn import datasets, linear_model, model_selection
```

## Load Dataset
Load the built-in diabetes dataset which includes 442 samples of data around diabetes with 10 feature variables.

1. Add some print commands to show the shape of dataset and its first element.

```python
# Load the built-in dataset to X
X, Y = datasets.load_diabetes(return_X_y=True)
print(X.shape)
print(X[0])
```

2. Select a portion of this dataset to plot.

```python
# Reshape this dataset to 2-D array
X = X[:, 2] 
X = X.reshape((-1, 1))
```

3. Now that you have data ready to be plotted, you can see if a machine can help split the training and testing dataset.

```python
# Split training, testing dataset by default function
X_train, X_test, Y_train, Y_test = model_selection.train_test_split(X, Y, train_size=0.8)
```

4. Then load up the linear regression model and train it with the above X_train, Y,train dataset.

```python
# Use Linear Regression model 
model = linear_model.LinearRegression()
model.fit(X_train, Y_train)
```

5. Then make a prediction using the test dataset.

```python
# Will need to compare the predict Y value with the actual test data
Y_pred = model.predict(X_test)
```

6. Show the data in a plot.

```python
plt.scatter(X_test, Y_test, color='black')
plt.plot(X_test, Y_pred, color='blue')
plt.xlabel("Scaled BMIs")
plt.ylabel("Disease Progression")
plt.title("Diabetes Progression against BMI")
plt.show()
```
